```python
import random

#Setup game
game = ["rock", "paper", "scissors"]

#Let computer choose at random and stores choice
computer_choice = random.choice(game)
print (computer_choice)

#Let user choose and store choice
user_choice = input("Enter rock, paper or scissors")

#Play game

print("Computer chose", computer_choice , "and you chose" , user_choice + "!")

if computer_choice == "rock":
    if user_choice == "rock":
        print("Tie")
    elif user_choice == "paper":
        print("User Wins!")
    else:
        print("Computer Wins")

if computer_choice == "paper":
    if user_choice == "paper":
        print("Tie")
    elif user_choice == "rock":
        print("User Wins!")
    else:
        print("Computer Wins")

if computer_choice == "scissors":
    if user_choice == "scissors":
        print("Tie")
    elif user_choice == "rock":
        print("User Wins!")
    else:
        print("Computer Wins")
```
